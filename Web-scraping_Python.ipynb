```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Step 1: 
## Extracting last 5 years tables (2019-2023) from wikipedia page of "Forbes list of the world's highest-paid athletes"

url = 'https://en.wikipedia.org/wiki/Forbes_list_of_the_world%27s_highest-paid_athletes'

page = requests.get(url)

soup = BeautifulSoup(page.text,'html.parser')

## Searching for all HTML 'table' elements on a webpage:
soup.find_all('table')

## Find the first table element as a BeautifulSoup object (regardless of class) :
soup.find_all('table')[0]

## Now get the first table element with more specific classes 'wikitable' and 'sortable'. 
## This is useful for situations where you're dealing with structured data on a Wikipedia page, for example.
Table0 = soup.find_all('table', class_='wikitable sortable')[0]

## find all the <th> elements within the table:
Table00 = Table0.find_all('th')

## Get the Headers:
Table00_Titles = [title.text.strip() for title in Table00]

## creating dataframe
df2023 = pd.DataFrame(columns = Table00_Titles)

## Getting table values for the dataframe:
column_data = Table0.find_all('tr')

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df2023)
    df2023.loc[length] = individual_row_data


## Add a new column 'Year' with value '2023' to all columns:
df2023 = df2023.assign(Year='2023')

## Rename the columns
df2023.rename(columns={'Salary/winnings': 'Salary'}, inplace=True)

df2023
```

```python
soup.find_all('table')[1]

Table1 = soup.find_all('table', class_='wikitable sortable')[1]

Table11 = Table1.find_all('th') #fetching the headings of the table

Table11_Titles = [title.text.strip() for title in Table11]

df2022 = pd.DataFrame(columns = Table11_Titles)

column_data1 = Table1.find_all('tr')

for row in column_data1[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df2022)
    df2022.loc[length] = individual_row_data

df2022 = df2022.assign(Year='2022')

df2022.rename(columns={'Salary/winnings': 'Salary'}, inplace=True)

df2022
```

```python
soup.find_all('table')[2]

Table2 = soup.find_all('table', class_='wikitable sortable')[2]

Table22 = Table2.find_all('th') #fetching the headings of the table

Table22_Titles = [title.text.strip() for title in Table22]

df2021 = pd.DataFrame(columns = Table22_Titles)

column_data2 = Table2.find_all('tr')

for row in column_data2[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df2021)
    df2021.loc[length] = individual_row_data

df2021 = df2021.assign(Year='2021')

df2021.rename(columns={'Salary/winnings': 'Salary'}, inplace=True)

df2021
```

```python
soup.find_all('table')[3]

Table3 = soup.find_all('table', class_='wikitable sortable')[3]

Table33 = Table3.find_all('th') #fetching the headings of the table

Table33_Titles = [title.text.strip() for title in Table33]

df2020 = pd.DataFrame(columns = Table33_Titles)

column_data3 = Table3.find_all('tr')

for row in column_data3[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df2020)
    df2020.loc[length] = individual_row_data

df2020 = df2020.assign(Year='2020')

df2020.rename(columns={'Salary/winnings': 'Salary'}, inplace=True)

df2020
```

```python
soup.find_all('table')[5]

Table5 = soup.find_all('table', class_='wikitable sortable')[5]

Table55 = Table5.find_all('th') #fetching the headings of the table

Table55_Titles = [title.text.strip() for title in Table55]

df2019 = pd.DataFrame(columns = Table55_Titles)

column_data4 = Table5.find_all('tr')

for row in column_data4[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df2019)
    df2019.loc[length] = individual_row_data

df2019 = df2019.assign(Year='2019')

## Rename the columns
df2019.rename(columns={'Nation': 'Country'}, inplace=True)
df2019.rename(columns={'Salary/Winnings': 'Salary'}, inplace=True)

df2019
```

```python
## Step 2:
## Combine the DataFrames
combined_df = pd.concat([df2023, df2022, df2021, df2020, df2019], ignore_index = True, join = 'inner')

print(combined_df)
```

```python
## Step 3:
## Convert the dataframe to CSV:
combined_df.to_csv(r'C:\Users\augas\Desktop\CANTEK Academy\Python\Highest-paid-athletes(2019-2023).csv', index = False)
```

```python
## Step 4:
## Data Cleaning & Transformation:

df = pd.DataFrame(combined_df)

df.rename(columns={'Total': 'Total in Millions', 'Salary': 'Salary in Millions','Endorsements': 'Endorsements in Millions'}, inplace=True)

# Remove '$' and 'millions' and convert to numeric type:
df['Total in Millions'] = pd.to_numeric(df['Total in Millions'].str.replace(r'[$,]', '', regex=True).str.replace(' million', '', regex=False))
df['Salary in Millions'] = pd.to_numeric(df['Salary in Millions'].str.replace(r'[$,]', '', regex=True).str.replace(' million', '', regex=False))
df['Endorsements in Millions'] = pd.to_numeric(df['Endorsements in Millions'].str.replace(r'[$,]', '', regex=True).str.replace(' million', '', regex=False))

df.info()
df
```

```python
## Step 5:
## Data Analysis and Visualization:

#Top Earners by Total Earnings:
top_earners = df.groupby('Name')['Total in Millions'].max().sort_values(ascending=False).head(10)
top_earners
```

```python
#Comparison of Sports Categories
sports_distribution = df.groupby('Sport')['Total in Millions'].sum().sort_values(ascending=False)
sports_distribution
```

```python
#Trends in Total Earnings Over the Years
earnings_over_years = df.groupby('Year')['Total in Millions'].sum()
earnings_over_years
```

```python
# Plotting Top Earners
plt.figure(figsize=(10, 6))
top_earners.plot(kind='bar', color='skyblue')
plt.title('Top Earners by Total Earnings')
plt.xlabel('Athlete')
plt.ylabel('Total Earnings (in Millions)')
plt.show()
```

```python
# Plotting Sports Distribution
plt.figure(figsize=(10, 6))
sports_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Earnings by Sport')
plt.ylabel('')
plt.show()
```

```python
# Plotting Trends in Total Earnings Over the Years
plt.figure(figsize=(10, 6))
earnings_over_years.plot(marker='o', color='green')
plt.title('Trends in Total Earnings Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Earnings (in Millions)')
plt.grid(True)
plt.show()
```
